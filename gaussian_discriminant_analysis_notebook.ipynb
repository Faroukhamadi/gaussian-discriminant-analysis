```python
import matplotlib.pyplot as plt
import numpy as np
import util

from gaussian_discriminant_analysis import GDA
```

## Dataset_1

*Note*: *`add_intercept` parameter includes an intercept term*

```python
x_train, y_train = util.load_dataset('data/ds1_train.csv', add_intercept=False)
x_valid, y_valid = util.load_dataset('data/ds1_valid.csv', add_intercept=True)
```

Take a look at the training set

```python
plt.xlabel('x1')
plt.ylabel('x2')
plt.plot(x_train[y_train == 1, 0], x_train[y_train == 1, 1], 'bx')
plt.plot(x_train[y_train == 0, 0], x_train[y_train == 0, 1], 'go')
```

Train the Gaussian Discriminant Analysis Model

```python
clf = GDA()
clf.fit(x_train, y_train)
preds = clf.predict(x_valid)
```

Plot the predictions for the training set

```python
util.plot(x_train, y_train, clf.theta)
print('Theta is: ', clf.theta)
x_train_with_intercept = np.c_[np.ones(x_train.shape[0]), x_train]
print('The accuracy of the training set is: ', np.mean(clf.predict(x_train_with_intercept) == y_train))
```

Plot the predictions for the validation set

```python
util.plot(x_valid, y_valid, clf.theta)
print('Theta is: ', clf.theta)
print('The accuracy of the validation set is: ', np.mean(clf.predict(x_valid) == y_valid))
```
